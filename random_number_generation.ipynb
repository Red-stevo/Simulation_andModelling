```python
from collections import Counter
from matplotlib import pyplot as plt


z0 = 19
a  = 22
m  = 63
m1 = 1000
c  = 4



random_number = []

for i in range(m1):
    z0 = (((a * z0) + c) % m) / m
    random_number.append(z0)

counter = Counter(random_number)

x_axis = range(m1)

plt.title("Linear Method For Random Number generation.")
plt.xlabel("Trials")
plt.ylabel("Random numbers.")

plt.plot(x_axis, random_number)
plt.show()
```

```python
for i in counter:
    print(f"{i} : {counter[i]}")


print(len(counter))

counter = None
```
